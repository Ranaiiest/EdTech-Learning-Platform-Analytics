```python
from pyspark.sql.functions import col, current_date, datediff, when
```

##### silver_students

```python
students_df = spark.read.table("bronze_students")
```

```python
students_df = students_df.withColumn(
    "registration_date",
    col("registration_date").cast("date")
)

students_df = students_df.withColumn(
    "account_age_days",
    datediff(current_date(), col("registration_date"))
)
```

```python
students_df.write.mode("overwrite").saveAsTable("silver_students")
```

```python
print(f"silver_students rows: {students_df.count()}")
```

##### silver_assessments

```python
assessments_df = spark.read.table("bronze_assessments")
```

```python
assessments_df = assessments_df.withColumn(
    "enrollment_date",
    col("enrollment_date").cast("date")
).withColumn(
    "completion_date",
    col("completion_date").cast("date")
)
```

```python
assessments_df = assessments_df.withColumn(
    "completion_days",
    when(
        col("completion_date").isNotNull(),
        datediff(col("completion_date"), col("enrollment_date"))
    )
)
```

```python
assessments_df = assessments_df.withColumn(
    "pass_flag",
    col("score") >= 60
)
```

```python
assessments_df.write.mode("overwrite").saveAsTable("silver_assessments")
```

```python
print(f"silver_assessments rows: {assessments_df.count()}")
```

##### silver_courses

```python
courses_df = spark.read.table("bronze_courses")
```

```python
courses_df = courses_df.withColumn(
    "revenue_per_enrollment",
    col("price_usd") * col("enrollment_count")
)
```

```python
courses_df.write.mode("overwrite").saveAsTable("silver_courses")
```

```python
print(f"silver_courses rows: {courses_df.count()}")
```
